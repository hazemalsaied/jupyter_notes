## Enrichment

```python
enriched_movies = pd.DataFrame()
for idx, row in movies.iterrows():
    search = movie.search(row['Name'])
    if search:
        added=False
        for ser in search: 
            if row['Name'] == ser.title:
                added = True
                response = requests.get(f'https://api.themoviedb.org/3/movie/{search[0].id}?api_key={tmdb.api_key}')
                if response.status_code == 200:
                    movie_details = response.json()
                    details = dict()
                    for c in ['adult',  'belongs_to_collection', 'budget',  'homepage', 'id', 'imdb_id', 'original_language', 'original_title', 'overview', 'popularity', 'poster_path', 
                               'release_date', 'revenue', 'runtime',  'status', 'tagline', 'title', 'video', 'vote_average', 'vote_count']:
                        details[c] =movie_details.get(c, '')
                    for c in ['genres', 'production_companies', 'production_countries', 'spoken_languages']:
                            key = 'english_name' if c == 'spoken_languages' else 'name'
                            details[c] = ' | '.join(sorted([e[key] for e in movie_details[c]]))


                    enriched_movies = enriched_movies.append({**row, **details}, ignore_index=True)
                print(f'The film {row["Name"]} was enriched successfully!')
                break
        if not added:
            print(row['Name'], '|'.join([r.title + getattr(r, 'release_date','    ')[:4] for r in search]))
            enriched_movies = enriched_movies.append({**row}, ignore_index=True)
```

```python
def rename(x):
    if x.startswith('Manoel’s Destinies'):
        return 'Manuel’s Destinies' + x[-3:]
    if x.startswith('The Decalogue'):
        return 'Le Décalogue' + x[-3:]
    if x.startswith('Neil Young'):
        return 'Year of the Horse'
    if x.startswith('Sunset Blvd Boulevard'):
        return 'Sunset Boulevard'
    if x.startswith('_The Class'):
        return 'The Class'
    if x.startswith('The Working Class Goes to Paradise'):
        return 'La classe operaia va in paradiso'
    if x.startswith('Welcome Back Mr.McDonald'):
        return 'Welcome Back, Mr. McDonald'
    if x.startswith('12.08 East of Bucharest'):
        return 'A fost sau n-a fost?'
    return x
non_enriched_movies['Name'] = non_enriched_movies['Name'].apply(lambda x: rename(x))
```
